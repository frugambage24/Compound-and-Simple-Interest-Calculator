```python
def simple_interest():
    amount = input("What is the amount of money you are provided?(Princal ")
    interest_rate = input("What is the interest rate that you pay back the amount? ")
    duration = input("Are you paying this money in duration of days, months, years? Please don't use any quotations just insert")

    try:
        amount = float(amount)
        interest_rate = float(interest_rate)
    except ValueError:
        return "Invalid input for amount or interest rate."

    if duration == "years":
        number_year = input("How many years? ")
        try:
            number_year = float(number_year)
        except ValueError:
            return "Invalid input for number of years. Note years should be in numerical figures"
        simple_interest = (amount * interest_rate * number_year) / 100
        return simple_interest
    elif duration == "months":
        number_months = input("How many months?")
        try:
            number_months = float(number_months)
        except ValueError:
            return "Invalid input for number of years. Note months should be in numerical figures"
        simple_interest = (amount * interest_rate * number_months) / (100 * 12)
        return simple_interest

    else:
        duration == "days"
        number_days = input("How many days? ")
        try:
            number_days = float(number_days)
        except ValueError:
            return "Invalid input for number of years. Note days should be in numberical figures"
        simple_interest = (amount * interest_rate * number_days)/ (100*366)
        return simple_interest
```
